# Create the population data

```python
pop2000_2010_path = "../data/cali_county_pop_2000_2010.csv"
pop2010_2020_path = "../data/cali_county_pop_2010_2020.csv" 

pop1 = pd.read_csv(pop2000_2010_path, header=0)
pop1.head()
pop1c = pop1.copy()
pop1c = pop1c.drop(index=pop1c.index[0])
pop1c.head()
```

```python
def clean_pop_dataframe(pop_data_path: str) -> pd.DataFrame:
  raw_data = pd.read_csv(pop_data_path)
  data_clean = raw_data.copy()
  data_clean = data_clean.drop(index=data_clean.index[0])
  data_clean['Geographic Area'] = data_clean['Geographic Area'].str.strip().str.replace(".", "", regex=False)
  return data_clean 
  
```

```python
pop1 = clean_pop_dataframe(pop2000_2010_path)
pop1.head()
```
