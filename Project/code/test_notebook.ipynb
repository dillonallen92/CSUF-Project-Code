```python
def clean_pop_dataframe(pop_data_path: str) -> pd.DataFrame:
    raw_data = pd.read_csv(pop_data_path)
    data_clean = raw_data.copy()
    data_clean = data_clean.drop(index=data_clean.index[0])# From current pop data, this is the row that just states the overall california population data, not useful for our cases right now
    data_clean['Geographic Area'] = data_clean['Geographic Area'].str.strip().str.replace(".", "", regex=False).str.replace(", California", "", regex=False) # Current first column values have a . before county name, i.e .Fresno
    return data_clean 
```
